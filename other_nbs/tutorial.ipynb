#### 4.1 Importing Libraries

```python
# Before setting up any script libraries need to be initialised to use most functions
# Here are some basic libraries to import
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### 4.2 Initializing variables and Paths

```python
root = r'C:\Users\Tyler\work\Flood_lab_notebooks\data' # the r before the address helps Python understand it's an address!!
line = np.arange(1,50,0.1) # Calling a function within the numpy library 
sin  = np.sin(line)
df   = pd.DataFrame()  # Creating a DataFrame using pandas
df.insert(0,'line',line)
df.insert(1,'sin',sin)
df.head()
```

#### 4.3 Plotting

```python
fig,ax = plt.subplots(figsize=(10,4))
ax.plot(line,sin,'or')
plt.xlabel('distance')
plt.ylabel('Waves')
plt.title('Example Sin Wave')
plt.show(fig)
```

#### 4.4 Plotting using dataframe

```python
fig,ax = plt.subplots(figsize=(10,4))
ax.plot(df['line'],df['sin'],'b')
plt.xlabel('distance')
plt.ylabel('Waves')
plt.title('Example Sin Wave')
plt.show(fig)
```

#### 4.5 Reading in a file

```python
file = 'water_level2.csv'
noaa = os.path.join(root,file) # using the os library I'm adding the file to the address
tide = pd.read_csv(noaa)       # pandas will read in the csv
tide.head(2)
```

#### 4.6 Plotting the data from file

```python
# plotting the data from the file is the same as earlier
#plt.close(fig)
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(tide['Date Time'],tide[' Water Level'])
plt.xlabel('Date time')
plt.ylabel('MSL (m)')
plt.title('Water Level (CBBT)')
plt.show()
```
