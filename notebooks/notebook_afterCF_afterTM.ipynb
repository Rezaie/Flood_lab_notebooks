### Original Project Description:
 - Run the iFlood SubX forecast system over a period for which SubX re-forecasts have been produced (1999-2015) to evaluate the overall predictability statistics of sea-level predictions at sites along the Chesapeake Bay where historical tide gauge data is available for validation, and the predictability at various lead times for individual events.
 - Evaluation of the predictability at the daily, weekly, up to 1-month timescale of the key variables, forecasted by the SubX models, that are needed to force the Mason iFlood system - namely 10-meter winds and sea-level pressure over the western North Atlantic. For this task we will look at the overall predictability statistics for the 1999-2015 re-forecast period over which the global weather and climate models participating in the SubX project have been run. We will evaluate the skill of the multi-model mean as well as the respective model ensembles.

---
### Testbed Description:
- Testbed using one ensemble member
- SubX Forcing using GEFS ensemble #10
- ADCIRC with EC95 mesh
- Model time step is 20s
- Model run time 35 days
- No tidal signals included
- Using the wind and pressure as a forcing the model is coldstarted
- Global outputs are generated every hour
- Station outputs are generated every 6 minutes
- Simulated time series of surges at NOAA coastal stations
- Observed Time Series of surge at NOAA coastal stations are shown with the simulated surges to be viewed as a reference.
- True is the Re-analysis Winds and Pressure used as forcing for the ADCIRC model
- Error metric #1: Storm surge magnitude bias comparing ensamble member vs control (time period considered 7 days prior/after landfall)

### Metric 1: Surge Bias Calculations
\begin{align}
Bias(dt) = Week_{peak}(dt) - Reanalysis_{peak}(dt)
\end{align}

- Where maximum peak is found within dt=14 days of the Hurricane making landfall
